```python
import pandas as pd
from pathlib import Path
import os
import shutil
```

```python

df = pd.read_csv(Path.cwd() / 'archive' / 'Respiratory_Sound_Database' / 'Respiratory_Sound_Database' / 'patient_diagnosis.csv')

df.groupby('URTI')

# keep only 5 values for each class
df = df.groupby('URTI').head(5)

print(df)

for index, row in df.iterrows():
    # copy all files from archive\Respiratory_Sound_Database\Respiratory_Sound_Database\audio_and_txt_files to archive\Respiratory_Sound_Database\Respiratory_Sound_Database\selected_files if they start with the same number
    for file in os.listdir(Path.cwd() / 'archive' / 'Respiratory_Sound_Database' / 'Respiratory_Sound_Database' / 'audio_and_txt_files'):
        if(str(row[0]) in file):
            # copy file in selected_files
            print(file)
            shutil.copy(Path.cwd() / 'archive' / 'Respiratory_Sound_Database' / 'Respiratory_Sound_Database' / 'audio_and_txt_files' / file, Path.cwd() / 'archive' / 'Respiratory_Sound_Database' / 'Respiratory_Sound_Database' / 'selected_files' / file)



    

```
